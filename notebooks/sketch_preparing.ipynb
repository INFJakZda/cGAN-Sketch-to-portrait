```python
def get_glob(data_set_path):
    return glob.glob(data_set_path + "/*/*/**.png")
    
photo_paths = get_glob("orginal_images_sample")
```

```python
def load_img(path):
    return io.imread(path)

def preprocess_img(img):
    img = color.rgba2rgb(img)[50:650,:,:]
    cropped_img = transform.resize(img, (256, 256, 3))
    return cropped_img
```

```python
def show_preprocessed_images():
    fig = plt.figure(figsize=(6, 30))
    cols, rows = 2, 15
    
    for i in range(1, rows + 1):
        fig.add_subplot(rows, cols, 2 * i - 1)
        img = load_img(photo_paths[i])
        plt.imshow(img)

        fig.add_subplot(rows, cols, 2 * i)
        img = preprocess_img(img)
        plt.imshow(img)

show_preprocessed_images()
```
